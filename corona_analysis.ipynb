```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
%matplotlib inline
```

```python
confirmed = pd.read_csv('data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
deaths = pd.read_csv('data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
recovered = pd.read_csv('data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv')
```

```python
confirmed.sum()[2:].plot()
plt.xlabel('date')
plt.ylabel('# of people')
plt.title('Confirmed infections')
plt.grid()
```

```python
data = confirmed.sum()[2:]
```

```python
x = data.pct_change()+1
```

```python
x.plot()
plt.title('Growth Factor')
plt.hlines(1,0,len(x),'r')
plt.ylim([0.5,2])
plt.grid()
```

Growth Factor between 0 and 1 means decline, growth factor above 1 indicates increase
